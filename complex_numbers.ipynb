```python
import matplotlib .pyplot as plt
import numpy as np
from matplotlib import style
z1=complex(5,7)
z2=complex(3,6)
z3=complex(4,2)
plt.figure(figsize=(12,9))
style.use('dark_background')
plt.plot(np.real(z1),np.imag(z1),'yo',markersize=15,label='(5+7j)')
plt.plot(np.real(z2),np.imag(z2),'mo',markersize=15,label='(3+6j)')
plt.plot(np.real(z3),np.imag(z3),'ro',markersize=15,label='(4+2j)')
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.plot([-10,10],[0,0],'g',lw=3)
plt.plot([0,0],[-10,10],'g',lw=3)
plt.plot([0,3],[0,6],'m-',lw=3)
plt.xlabel('Real axis',fontsize=20)
plt.ylabel('Imaginary axis',fontsize=20)
plt.legend(fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
t=np.linspace(0,10,100)
f=50
A=1
theta=np.pi/2
complex_wave=np.exp(1j*(2*np.pi*f*t+theta))
plt.figure(figsize=(16,10))
style.use('dark_background')
plt.plot(t,np.real(complex_wave),'m-',linewidth=3,label='real-part')
plt.plot(t,np.imag(complex_wave),'b-',linewidth=3,label='imag-part')
plt.xlim([-0.1,10])
plt.xlabel('time',fontsize=15)
plt.ylabel('Amplitude',fontsize=15)
plt.title('Complex- sinewave',fontsize=20)
plt.legend(fontsize=15)
plt.show()
```
