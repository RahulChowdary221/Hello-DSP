```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
srate=256
amp=3
f=5 #hz
t=np.arange(0,1,1/srate)
x=amp*np.sin(2*np.pi*f*t)
X=np.zeros(len(x),dtype=complex)
for freq in range(0,len(t)):
 csw=np.exp(-1j*2*np.pi*freq*t)
 X[freq]=np.sum(x*csw)
amps=np.abs(X)/len(t);

```

```python
plt.figure(figsize=(24,10))
style.use('dark_background')
plt.rcParams['xtick.labelsize']=15
plt.rcParams['xtick.labelsize']=15
plt.plot(t,3*np.real(csw),'r-',linewidth=3,label='real part of csw')
plt.plot(t,3*np.imag(csw),'m-',linewidth=3,label='imag part of csw')
plt.plot(t,x,'w-',linewidth=3,label='sine-wave')
plt.xlabel('time in sec',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)
plt.title('Signal-Modelling using Complex sine-wave',fontsize=24)
plt.xlim([0,1])
plt.legend(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(24,10))
style.use('dark_background')
plt.rcParams['xtick.labelsize']=15
plt.rcParams['xtick.labelsize']=15
plt.subplot(2,2,1)
plt.plot(t,x,'r-',linewidth=3,label='sine-wave')
plt.xlabel('time in sec',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)
plt.title('Sine-wave',fontsize=24)
plt.subplot(2,2,2)
markerline,stemlines,baseline=plt.stem(amps)
plt.setp(stemlines,'linewidth',3)
plt.xlabel('Indices',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)
plt.title('Fourier Transform of a signal',fontsize=24)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
srate=256
amp=3
f=5 #hz
t=np.arange(0,1,1/srate)
x=amp*np.sin(2*np.pi*f*t)
X=np.zeros(len(x),dtype=complex)
for freq in range(0,len(t)):
 csw=np.exp(-1j*2*np.pi*freq*t)
 X[freq]=np.sum(x*csw)
amps=2*np.abs(X)/len(t); #multipliting by 2 to incorporate negative frequencies on complex sinusoids
```

```python
#converting indices to frquencies
#There are 0 to 255 indices at which the sinewave .These are frequencies indices
#not the frequencies in hertz,Since we have n=256,indices including zero,ie n/2=128 we get.. 
#negate waveforms of first 128 samples.it means after n=128,
#Thus frequencis are violating the nyquist criteria
#Thus inorder to convert indices into hertz we need (n/2+1) or (n/2)linearly spaced btw zero
#Nyquist (srate/2) as shown 
import math 
Nyquist =srate/2
Hz=np.linspace(0,Nyquist,math.floor(len(t)/2 +1))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
srate=256
t=np.arange(0,1,1/srate)
x1=5*np.sin(2*np.pi*2*t)
x2=2*np.sin(2*np.pi*5*t)
x3=7*np.sin(2*np.pi*7*t)
x4=x1+x2+x3
X=np.zeros(len(x4),dtype=complex)
for freq in range(0,len(t)):
    csw=np.exp(-1j*2*np.pi*freq*t)
    X[freq]=np.sum(x4*csw)
amps=2*np.abs(X)/len(t)
Nyquist =srate/2
Hz1=np.linspace(0,Nyquist,math.floor(len(t)/2 +1))
plt.figure(figsize=(20,10))
plt.rcParams['xtick.labelsize']=20
plt.rcParams['ytick.labelsize']=20
plt.plot(t,x4,'r-',linewidth=3,label='Combined sinewaves')
plt.plot(t,np.real(csw),'b-',linewidth=3,label='Real-part of csw')
plt.plot(t,np.imag(csw),'c-',linewidth=3,label='Imag-part of csw')
plt.xlabel('time in sec',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)
plt.legend(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(24,14))
style.use('dark_background')
plt.rcParams['xtick.labelsize']=15
plt.rcParams['xtick.labelsize']=15
plt.subplot(2,2,1)
plt.plot(t,x4,'r-',linewidth=3,label='sine-wave')
plt.xlabel('time in sec',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)
plt.title('signal with frequencies 2 5 and 7Hz',fontsize=24)
plt.subplot(2,2,2)
markerline,stemlines,baseline=plt.stem(Hz1,amps[range(0,len(Hz1))])
plt.setp(stemlines,'linewidth',3)
plt.xlim(0,10)
plt.ylim(0,8)
plt.xlabel('frequency in herts',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)
plt.title('Fourier Transform of a signal',fontsize=24)
plt.show()  
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import scipy.fft
import math
srate=256
Nyquist=srate/2
t=np.arange(0,1,1/srate) #time vector in seconds
x1=5*np.sin(2*np.pi*2*t)
x2=2*np.sin(2*np.pi*4*t)
x3=7*np.sin(2*np.pi*6*t)
x=x1+x2+x3 #combined sinewaves
#calculating Fourier-coefficient
X=scipy.fft.fft(x)/len(t) 
X=2*np.abs(X)
#converting indices into frequency
Hz2=np.linspace(0,Nyquist,math.floor(len(t)/2+1))
plt.figure(figsize=(24,14))
plt.suptitle('Signal and its Fast-Fourier Transform')
style.use('dark_background')
plt.rcParams['xtick.labelsize']=25
plt.rcParams['ytick.labelsize']=25
plt.subplot(2,1,1)
plt.plot(t,x,'r-',label='Sinewave with frequencies 2,4,6')
plt.title('Combined sinewave')
plt.subplot(2,1,2)
markerline, stemlines, baseline = plt.stem(Hz1,amps[range(0,len(Hz1))])
plt.setp(stemlines,linewidth=2)
plt.setp(stemlines,color='cyan')
plt.setp(markerline, markersize=8, markerfacecolor='yellow')
plt.xlim(0,10)
plt.ylim(0,max(x)*1.1)
plt.xlabel('frequency in hertzs',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)
plt.title('Fourier Transform of a signal',fontsize=20)
plt.grid(True)
plt.tight_layout()
plt.legend(fontsize=15)
plt.show() 
```
