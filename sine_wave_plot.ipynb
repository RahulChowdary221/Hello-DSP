```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
x1=np.linspace(0,10,100)
y1 = x1**2
y2 = x1**3
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.subplot(2,2,2)
plt.subplot(2,2,2)
plt.plot(x1,y2)
plt.tight_layout()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
t = np.linspace(0,10,100)
A= 5
f =100
sine_wave = A*np.sin(2*np.pi*f*t)
plt.figure(figsize=(16,8))
style.use('dark_background')
plt.subplot(2,1,1)
plt.plot(t,sine_wave,'r')
plt.subplot(2,1,2)
plt.stem(t,sine_wave,'m')
plt.show()
```
