```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
t = np.linspace(0,10,100)
A= 5
f =100
sine_wave = A*np.sin(2*np.pi*f*t)
plt.figure(figsize=(16,8))
style.use('dark_background')
plt.subplot(2,1,1)
plt.plot(t,sine_wave,'r')
plt.subplot(2,1,2)
plt.stem(t,sine_wave,'m')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
t =np.arange(0,1,1/300)
x1=5*np.sin(2*np.pi*2*t)
x2=7*np.sin(2*np.pi*2*t)
x3=3*np.sin(2*np.pi*6*t)
x4=x1+x2+x3
plt.figure(figsize=(24,14))
style.use('dark_background')
plt.suptitle("Construction of different Sine-waves",fontsize =40) #super-title
plt.subplot(2,2,1)
plt.plot(t,x1,'g',lw=3)
plt.title("First Sine_wave")
plt.xlabel("time in sec",fontsize=20)
plt.ylabel("Amplitude",fontsize=20)
plt.subplot(2,2,2)
plt.plot(t,x2,'r',lw=3)
plt.title("Second Sine_wave")
plt.xlabel("time in sec",fontsize=20)
plt.ylabel("Amplitude",fontsize=20)
plt.subplot(2,2,3)
plt.plot(t,x3,'p',lw=3)
plt.title("Third Sine_wave")
plt.xlabel("time in sec",fontsize =20)
plt.ylabel("Amplitude",fontsize = 20)
plt.subplot(2,2,4)
plt.plot(t,x4,'m',lw=3)
plt.title("fourth Sine_wave")
plt.xlabel("time in sec",fontsize=20)
plt.ylabel("Amplitude",fontsize=20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
t =np.arange(0,1,1/300)
x1=2*np.sin(2*np.pi*2*t)
x2=5*np.sin(5*np.pi*5*t)
x3=7*np.cos(7*np.pi*7*t)
x4=1/2*np.pi*(x1+x2+x3)
plt.figure(figsize=(24,14))
plt.suptitle("Construction of different wave forms",fontsize=40)
style.use('dark_background')
plt.rcParams['lines.linewidth']=3
plt.rcParams['lines.linestyle']='--'
plt.rcParams['xtick.labelsize']=25
plt.rcParams['ytick.labelsize']=25
plt.subplot(2,2,1)
plt.plot(t,x1,'r')
plt.title("First-Sine Wave")
plt.subplot(2,2,2)
plt.plot(t,x2,'y')
plt.title("Second sine wave")
plt.subplot(2,2,3)
plt.plot(t,x3,'g')
plt.title("Third-Cos Wave")
plt.subplot(2,2,4)
plt.plot(t,x4,'m')
plt.title("Fourth-sine wave")
plt.show()
```
